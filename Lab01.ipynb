```python
import matplotlib.pyplot as plt

exp_samples = np.random.exponential(scale=1/lambda_value, size=100000)

uniform_samples = np.random.rand(100000)
normal_samples = np.random.randn(100000)

# Plot histograms
plt.figure(figsize=(12, 6))

plt.hist(exp_samples, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_samples, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_samples, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_title('3D Plot of Z = X^2 + Y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display the first few rows to understand the data
print(df.head())


features = df[['Defense', 'Speed', 'Stage', 'Total', 'HP', 'Attack']]

# Drop rows with missing values to ensure the correlation calculations are valid
features = features.dropna()

# Calculate Pearson correlation
pearson_corr = features.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = features.corr(method='spearman')

# Plot heatmaps
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Heatmap')

plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files


uploaded = files.upload()


file_name = list(uploaded.keys())[0]

# Load the dataset
df = pd.read_csv(file_name)

# Create a new dataframe by filtering the specified columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Display the first 20 samples of the new dataframe
print("First 20 samples of the new dataframe:")
print(new_df.head(20))

# Calculate mean, standard deviation, and quartiles
mean_values = new_df.mean()
std_values = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

# Display the results
print("\nMean Values:\n", mean_values)
print("\nStandard Deviation:\n", std_values)
print("\nQuartile Values:\n", quartiles)

# Plot histograms of each feature in a 1x4 layout
plt.figure(figsize=(20, 5))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[column], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Visualize the scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()

# Create a PairGrid with customized plots
g = sns.PairGrid(new_df)

# Plots on the diagonal: histogram with KDE
g.map_diag(sns.histplot, kde=True)



# Plots on the upper triangle: scatter plots
g.map_upper(sns.scatterplot)

plt.show()
```
